# Planetary nebulae with CLOUDY

```python
import numpy as np
import matplotlib.pyplot as plt

r, T = np.loadtxt("planetary_nebulapn.ovr", skiprows=1, usecols=(0, 1), unpack=True)

plt.plot(r, np.log10(T))
plt.xlabel(r'Radius [cm]')
plt.ylabel(r'logT [K]')
```

```python
nu, total_emission = np.loadtxt("planetary_nebulapn.con", skiprows=1, usecols=(0, 6), unpack=True)


plt.plot(np.log10(nu), np.log10(total_emission))
plt.xlabel(r'$log\nu [\mu]$')
plt.ylabel(r'$Total emission L_\nu [erg s^{-1}]$')
```
